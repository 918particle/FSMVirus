```python
class FindAverage:
    
    def __init__(self,nFiles,n):
        self.averages_y = list(0 for i in range(n))
        self.averages_z = list(0 for i in range(n))
        self.n_files = nFiles
    
    def average(self,fileList):
        data=[]
        i=0
        while i<len(fileList):
            f=open(fileList[i],'r')
            x_data = []
            y_data = []
            z_data = []
            for line in f:
                line = line.rstrip("\n")
                [x,y,z] = line.split(' ')
                x_data.append(int(x))
                y_data.append(int(y))
                #z_data.append(int(z))            
            if(i==0):
                data.append(x_data)
            data.append(y_data)
            f.close()
            i+=1
        n = len(data[0])
        m = len(fileList)+1
        for j in range(1,m):
            for i in range(n):
                self.averages_y[i] += float(data[j][i])/float(m-1)
        return self.averages_y
    
    def organizeResults(self,fileTitle):
        f = open(fileTitle,'w')
        for i in self.averages_y:
            f.write(str(i)+'\n')
        f.close()
```

```python
files = ['Oct31_run_0.dat','Oct31_run_1.dat','Oct31_run_2.dat']
the_averager = FindAverage(3,500)
print(the_averager.average(files))
the_averager.organizeResults('output.dat')
```
